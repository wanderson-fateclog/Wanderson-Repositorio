```python
import matplotlib.pyplot as plt
```

```python
exp_final['DATA'] = pd.to_datetime(exp_final['CO_ANO'].astype(str) + '-' + exp_final['CO_MES'].astype(str) + '-01')
```

```python
filtro_periodo = (exp_final['DATA'] >= '2025-01-01') & (exp_final['DATA'] <= '2026-01-01')
df_filtrado = exp_final[filtro_periodo]
```

```python
evolucao_kg = df_filtrado.groupby('DATA')['KG_LIQUIDO'].sum()
```

```python
plt.figure(figsize=(12, 6))
evolucao_kg.plot(kind='line', marker='o', color='navy', linewidth=2)

plt.title('Evolução do Peso Líquido Exportado (Jan/2025 - Jan/2026)', fontsize=14)
plt.xlabel('Mês de Referência', fontsize=12)
plt.ylabel('Total KG Líquido', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
